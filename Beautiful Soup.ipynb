```python

allPages = ['http://books.toscrape.com/catalogue/page-1.html',
            'http://books.toscrape.com/catalogue/page-2.html']
base_url='http://books.toscrape.com/catalogue/'
url=[]
column_names = ['Title', 'Link', 'Price', 'Quantity in Stock']
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd

title=[]
price=[]
qty=[]
for i in allPages:
    book_url=[]
    response=requests.get(i)
    data=BeautifulSoup(response.text,'html.parser')
    a=data.find_all(class_='product_pod')
    for i in a:
        book_url.append(base_url+i.a['href'])
    
    for i in book_url:
        url.append(i)
        res=requests.get(i)
        book_data=BeautifulSoup(res.text,'html.parser')
        title.append(book_data.h1.string)
        price.append(book_data.find(class_='price_color').string)
        qty.append(book_data.find(class_='instock availability').contents[-1].strip())
        
        
#print(title)
#print(book_url)
p=[]
for i in price:
    p.append(float(re.search('[\d.]+',i).group()))
##print(p)
quantity=[]
for i in qty:
    quantity.append(int(re.search('\d+',i).group()))
##print(quantity)

book_details=[]
for k in range(len(p)):
    book_details.append([title[k],url[k],p[k],quantity[k]])
df=pd.DataFrame(book_details,columns=column_names)
d=df.copy()
print(d)
#df.to_csv('Books.csv',index=False)
```

```python
import requests
from bs4 import BeautifulSoup
dict=[]
for i in range(1,11):
    response=requests.get('http://quotes.toscrape.com/page/'+str(i)+'/')
    html_data=response.text
    data=BeautifulSoup(html_data,'html.parser')
    li=data.find_all(class_="author")
    for i in li:
        if (i.string) in dict:
            continue
        elif i.string[0]=='J':
            dict.append(i.string)
dict.sort()
print(dict)
a={}
for key in dict:
    r=requests.get('http://quotes.toscrape.com/author/'+key+'/')
    html_data=r.text
    data=BeautifulSoup(html_data,'html.parser')
    li=data.select(".author-born-date")
    for i in li:
        a[key]=i.string
print(a)
    
```
