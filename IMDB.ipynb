```python
import requests
from bs4 import BeautifulSoup
basic='http://quotes.toscrape.com/page/'
for i in range(1,11):
    res=requests.get(basic+str(i)+'/')
    data=BeautifulSoup(res.text,'html.parser')
    a=data.find_all(class_='quote')
    for i in a:
        if i.find(class_='author').string=='Albert Einstein':
            print(i.find(class_='text').string)
```
